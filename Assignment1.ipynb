```python
#Question 4a and Question 4b
import math
x = [1, 2, 4, -4]
y = [10.5, 12.1, 13.2, 105.8]
def distance(x, y):
    n = len(x)
    d = 0.0
    for k in range(n):
        d += math.pow(x[k] - y[k],2)
    d = math.sqrt(d)
    return d
answer = distance(x, y)
print("The distance between the point(s) is",answer)
```

```python
#Question 6c
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/parkinsons.data.txt')

plt.hist(df['DFA'])
plt.title('DFA')
plt.show()

plt.hist(df['PPE'])
plt.title('PPE')
plt.show()

plt.hist(df['MDVP:Fo(Hz)'])
plt.title('MDVP:Fo(Hz)')
plt.show()



```

```python
#Question 6d
import pandas as pd
import matplotlib.pyplot as plt

status_1 = df[df['status'] == 1]
status_0 = df[df['status'] == 0]

attributes = ['PPE', 'DFA', 'MDVP:Fo(Hz)']

for attribute in attributes:
    plt.figure(figsize=(10, 5))
    plt.hist(status_1[attribute], alpha=0.5, label='Status 1', color='blue')
    plt.hist(status_0[attribute], alpha=0.5, label='Status 0', color='red')
    plt.xlabel(attribute)
    plt.ylabel('Frequency')
    plt.title('Histogram of {} for Different Health Status'.format(attribute))
    plt.legend()
    plt.show()
```
